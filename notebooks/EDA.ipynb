```python
import pandas as pd
import plotly.express as px

# Funcin para cargar los datos
def cargar_datos():
    return pd.read_csv('vehicles_us.csv')

# Funcin para crear el histograma
def crear_histograma(data, columna):
    return px.histogram(data, x=columna)

# Cargar los datos
car_data = cargar_datos()

# Crear el histograma y mostrarlo
fig = crear_histograma(car_data, "odometer")
fig.show()  # Mostrar el grfico en el cuaderno
```
